```python
import os
import uuid
img = os.path.join ('data', 'images', 'B.b6059d93-db97-11ec-80a6-dca266317fe4.jpg')
```

```python
%matplotlib inline
plt.imshow(np.squeeze(result.render()))
plt.show()
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    results = model(frame)
    
    cv2.imshow ('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
    
```

```python
cap.release()
cv2.destroyAllWindows()
```
